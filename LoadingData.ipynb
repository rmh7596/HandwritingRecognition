## Dimensions
Image must be 1280x1628
That makes each cell 160x148

```python
import os
import cv2
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image
trainPath = "Train/"
currentLetter = "o"
trainLetterPath = "Train/" + currentLetter
```

```python
image = cv2.imread(trainPath + "/" + currentLetter + ".JPG")
conv_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
rows = conv_image.shape[0]
cols = conv_image.shape[1]
b = np.lib.stride_tricks.as_strided(conv_image, shape=(11,8,148,160), strides=(189440, 160, 1280, 1))
pixelsPerInch = int(cols / 8.5)
print(conv_image.shape)
print(conv_image.strides)
print(b.shape)
print(b.strides)
plt.imshow(b[0][3], cmap='gray')
plt.show()
```

```python
iterations = 0
for i in range (11):
    for j in range(8):
        newArray = b[i][j]
        newImage = Image.fromarray(newArray)
        fileName = currentLetter + str(iterations) + '.jpg'
        newImage.save(f"{trainLetterPath}/{fileName}")
        iterations+=1
```
