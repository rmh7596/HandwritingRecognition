```python
import os
import cv2
import matplotlib.pyplot as plt
from PIL import Image
trainPath = "Train/"
trainAPath = "Train/a"
```

```python
image = cv2.imread(trainPath + "/DSC_0647.JPG")
conv_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
cropped_image = conv_image[180:1800, 1050:2225]
rows = cropped_image.shape[0]
cols = cropped_image.shape[1]
pixelsPerInch = int(cols / 8.5) # Width of paper
plt.imshow(cropped_image, cmap='gray')
plt.show()
```

```python
iterations = 0
for i in range(0, 11, pixelsPerInch):
    for j in range(0, 8, pixelsPerInch):
        newArray = cropped_image[j:j+pixelsPerInch, i:i+pixelsPerInch]
        
        newImage = Image.fromarray(newArray)
        fileName = 'a' + str(iterations) + '.jpg'
        newImage.save(f"{trainAPath}/{fileName}")
        iterations+=1
```
