## Data Augmentation
To avoid handwriting more characters, I will augment the current data I have to produce more images

```python
import matplotlib.pyplot as plt
import os
import PIL
from PIL import Image
```

```python
a = "Data/a/a0.jpg"

img1 = Image.open(a)
topRightQuad = img1.crop((50,0,160,120))
topLeftQuad = img1.crop((0,0,115,115))
bottomLeftQuad = img1.crop((0,50,115,148))
bottomRightQuad = img1.crop((50,60,160,148))

topRightCropped = topRightQuad.resize((160,148), resample=PIL.Image.BILINEAR)
topLeftCropped = topLeftQuad.resize((160,148), resample=PIL.Image.BILINEAR)
bottomLeftCropped = bottomLeftQuad.resize((160,148), resample=PIL.Image.BILINEAR)
bottomRightCropped = bottomRightQuad.resize((160,148), resample=PIL.Image.BILINEAR)

plt.imshow(img1, cmap="gray")
plt.show()
plt.imshow(topRightCropped, cmap="gray")
plt.show()
plt.imshow(topLeftCropped, cmap="gray")
plt.show()
plt.imshow(bottomLeftCropped, cmap="gray")
plt.show()
plt.imshow(bottomRightCropped, cmap="gray")
plt.show()
```

```python
letter = ""
letterDir = "Data/" + letter
labelNum = 87
for sample in os.listdir(letterDir):
    image = Image.open(os.path.join(letterDir,sample))
    topRightQuad = image.crop((50,0,160,120))
    topLeftQuad = image.crop((0,0,115,115))
    bottomLeftQuad = image.crop((0,50,115,148))
    bottomRightQuad = image.crop((50,60,160,148))

    topRightCropped = topRightQuad.resize((160,148), resample=PIL.Image.BILINEAR)
    topLeftCropped = topLeftQuad.resize((160,148), resample=PIL.Image.BILINEAR)
    bottomLeftCropped = bottomLeftQuad.resize((160,148), resample=PIL.Image.BILINEAR)
    bottomRightCropped = bottomRightQuad.resize((160,148), resample=PIL.Image.BILINEAR)

    labelNum+=1
    fileName = letter + str(labelNum) + '.jpg'
    topLeftCropped.save(f"{letterDir}/{fileName}")
    
    labelNum+=1
    fileName = letter + str(labelNum) + '.jpg'
    topRightCropped.save(f"{letterDir}/{fileName}")

    labelNum+=1
    fileName = letter + str(labelNum) + '.jpg'
    bottomLeftCropped.save(f"{letterDir}/{fileName}")

    labelNum+=1
    fileName = letter + str(labelNum) + '.jpg'
    bottomRightCropped.save(f"{letterDir}/{fileName}")
        
```
